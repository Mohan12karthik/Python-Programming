```python
def funcUpdation(currPassword, newPassword):
    len_curr = len(currPassword)
    len_new = len(newPassword)
    
    # Create a table to store lengths of longest common subsequence.
    dp = [[0] * (len_new + 1) for _ in range(len_curr + 1)]
    
    # Fill the table to find the length of the LCS
    for i in range(1, len_curr + 1):
        for j in range(1, len_new + 1):
            if currPassword[i - 1] == newPassword[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # Length of the longest common subsequence
    lcs_length = dp[len_curr][len_new]
    
    # Calculate the number of deletions and insertions
    deletions = len_curr - lcs_length
    insertions = len_new - lcs_length
    
    # Total number of operations
    return deletions + insertions

def main():
    import sys
#     input = sys.stdin.read
#     data = input().strip().split('\n')
    
#     currPassword = data[0].strip()
#     newPassword = data[1].strip()
    
#     result = funcUpdation(currPassword, newPassword)
    
#     print(result)
    currPassword = input("Enter the current password: ").strip()
    newPassword = input("Enter the new password: ").strip()
    
    # Call the function to compute the result
    result = funcUpdation(currPassword, newPassword)
    
    # Print the result
    print("Number of operations needed:", result)

if __name__ == "__main__":
    main()
```
