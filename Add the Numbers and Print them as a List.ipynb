```python
def funcSum(list1, list2):
    # Convert lists to numbers by reversing and joining the digits
    num1 = int(''.join(map(str, reversed(list1))))
    num2 = int(''.join(map(str, reversed(list2))))
    
    # Calculate the sum of the numbers
    total_sum = num1 + num2
    
    # Convert the sum to a string and reverse it
    reversed_sum_str = str(total_sum)[::-1]
    
    # Return the reversed sum as space-separated digits
    return ' '.join(reversed_sum_str)

def main():
    import sys
#     input = sys.stdin.read
#     data = input().split()
    
#     # Read sizes and lists from input
#     list1_size = int(data[0])
#     list1 = list(map(int, data[1:list1_size + 1]))
#     list2_size = int(data[list1_size + 1])
#     list2 = list(map(int, data[list1_size + 2:list1_size + 2 + list2_size]))
    
#     # Compute result
#     result = funcSum(list1, list2)
    
#     # Print result
#     print(result)

        # Prompt the user for the size and elements of the first list
    list1_size = int(input("Enter the size of the first list: "))
    list1 = list(map(int, input(f"Enter {list1_size} elements of the first list separated by space: ").split()))
    
    # Prompt the user for the size and elements of the second list
    list2_size = int(input("Enter the size of the second list: "))
    list2 = list(map(int, input(f"Enter {list2_size} elements of the second list separated by space: ").split()))
    
    # Compute result
    result = funcSum(list1, list2)
    
    # Print result
    print("Resulting sum in reverse order: ", result)

if __name__ == "__main__":
    main()
```
