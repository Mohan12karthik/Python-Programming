```python
def maxSum(inputArr):
    N = len(inputArr)
    
    if N == 0:
        return 0
    if N == 1:
        return inputArr[0]
    
    # Create a dp array
    dp = [0] * N
    
    # Base cases
    dp[0] = inputArr[0]
    dp[1] = max(inputArr[0], inputArr[1])
    
    # Fill the dp array based on the recurrence relation
    for i in range(2, N):
        dp[i] = max(dp[i-1], inputArr[i] + dp[i-2])
    
    # The last element of dp contains the maximum chocolates that can be collected
    return dp[N-1]

def main():
    inputArr_size = int(input("Enter the size of the array: "))  # Taking the size of the array
    inputArr = list(map(int, input("Enter the elements of the array: ").split()))  # Taking the array elements
    
    result = maxSum(inputArr)
    print(f"The maximum sum of chocolates is: {result}")

if __name__ == "__main__":
    main()
```
