```python
def funcMeanMode(inputArr):
    from collections import Counter
    
    N = len(inputArr)
    
    # Compute the mean
    total_sum = sum(inputArr)
    mean = total_sum // N
    
    # Compute the mode
    counts = Counter(inputArr)
    max_frequency = max(counts.values())
    
    # Get all numbers with the maximum frequency
    mode_candidates = [num for num, freq in counts.items() if freq == max_frequency]
    
    # For mode, return the smallest number in case of ties
    mode = min(mode_candidates)
    
    return [mean, mode]

def main():
    import sys
#     input = sys.stdin.read
#     data = input().split()
    
#     inputArr_size = int(data[0])
#     inputArr = list(map(int, data[1:]))
    
#     result = funcMeanMode(inputArr)
#     print(" ".join(map(str, result)))
    # Prompt the user for the size of the array
    inputArr_size = int(input("Enter the size of the array: "))
    
    # Prompt the user for the elements of the array
    inputArr = list(map(int, input(f"Enter {inputArr_size} elements separated by space: ").split()))
    
    result = funcMeanMode(inputArr)
    print("Mean and Mode: ", " ".join(map(str, result)))

if __name__ == "__main__":
    main()
```
