```python
def funcMatrix(matrix):
    # Step 1: Find the largest elements in each row
    largest_in_row = {}
    for i in range(len(matrix)):
        max_element = max(matrix[i])
        max_indices = [j for j, x in enumerate(matrix[i]) if x == max_element]
        for j in max_indices:
            largest_in_row[(i, j)] = max_element
    
    # Step 2: Check if these largest elements are the smallest in their columns
    for (i, j), value in largest_in_row.items():
        is_smallest_in_column = True
        for row in matrix:
            if row[j] < value:
                is_smallest_in_column = False
                break
        if is_smallest_in_column:
            return value
    
    return -1

def main():
    import sys
#     input = sys.stdin.read
#     data = input().strip().split()
    
#     matrix_rows = int(data[0])
#     matrix_cols = int(data[1])
    
#     matrix = []
#     index = 2
#     for _ in range(matrix_rows):
#         row = list(map(int, data[index:index + matrix_cols]))
#         matrix.append(row)
#         index += matrix_cols
    
#     result = funcMatrix(matrix)
#     print(result)

    # Prompting user for input
    matrix_rows = int(input("Enter the number of rows in the matrix: "))
    matrix_cols = int(input("Enter the number of columns in the matrix: "))

    # Reading matrix elements from the user
    matrix = []
    print(f"Enter the elements of the matrix, row by row:")
    for _ in range(matrix_rows):
        row = list(map(int, input().split()))
        if len(row) != matrix_cols:
            print(f"Error: Please enter exactly {matrix_cols} elements for each row.")
            return
        matrix.append(row)
    
    result = funcMatrix(matrix)
    print(result)

if __name__ == "__main__":
    main()
```
