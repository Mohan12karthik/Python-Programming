```python
def funcMatrix(matrix):
    if not matrix:
        return -1
    
    matrix_rows = len(matrix)
    matrix_cols = len(matrix[0])
    
    # Step 1: Find the largest elements in each row and their column indices
    largest_in_row = []
    for i in range(matrix_rows):
        max_value = matrix[i][0]
        max_col = 0
        for j in range(matrix_cols):
            if matrix[i][j] > max_value:
                max_value = matrix[i][j]
                max_col = j
        largest_in_row.append((max_value, max_col))
    
    # Step 2: Check if the largest element in each row is the smallest in its column
    for value, col in largest_in_row:
        is_smallest_in_col = True
        for i in range(matrix_rows):
            if matrix[i][col] < value:
                is_smallest_in_col = False
                break
        if is_smallest_in_col:
            return value
    
    return -1

def main():
    import sys
#     input = sys.stdin.read
#     data = input().strip().split()
    
#     matrix_rows = int(data[0])
#     matrix_cols = int(data[1])
    
#     matrix = []
#     index = 2
#     for _ in range(matrix_rows):
#         matrix.append([int(data[i]) for i in range(index, index + matrix_cols)])
#         index += matrix_cols
    
#     result = funcMatrix(matrix)
#     print(result)

    # Prompt the user for the matrix dimensions
    matrix_rows = int(input("Enter the number of rows: "))
    matrix_cols = int(input("Enter the number of columns: "))
    
    # Initialize the matrix
    matrix = []
    
    # Prompt the user to enter the matrix row by row
    print("Enter the matrix row by row:")
    for i in range(matrix_rows):
        row = list(map(int, input().split()))
        if len(row) != matrix_cols:
            print(f"Error: You must enter exactly {matrix_cols} numbers for each row.")
            return
        matrix.append(row)
    
    # Call the function to find the desired value
    result = funcMatrix(matrix)
    print("Result:", result)

if __name__ == "__main__":
    main()
```
