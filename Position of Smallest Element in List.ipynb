```python
def funcSmallest(inputList):
    # Initialize the minimum value and its position
    if not inputList:
        return -1  # Return -1 if the list is empty (although not specified, it's good practice)
    
    min_value = inputList[0]
    min_index = 0
    
    # Traverse the list to find the minimum value and its index
    for index in range(1, len(inputList)):
        if inputList[index] < min_value:
            min_value = inputList[index]
            min_index = index
            
    return min_index

def main():
    import sys
#     input = sys.stdin.read
#     data = input().strip().split()
    
#     # Read the size of the list
#     inputList_size = int(data[0])
    
#     # Read the elements of the list
#     inputList = list(map(int, data[1:1 + inputList_size]))
    
#     # Get the position of the smallest number
#     result = funcSmallest(inputList)
    
#     # Print the result
#     print(result)

    # Prompt the user for input
    input_data = input("Enter the size of the list followed by the elements, separated by spaces: ")
    data = input_data.strip().split()
    
    # Check if we have at least one element for size
    if len(data) == 0:
        print("Error: No input provided.")
        return

    # Read the size of the list
    inputList_size = int(data[0])
    
    # Check if the list has the correct number of elements
    if len(data) < 1 + inputList_size:
        print("Error: The number of elements does not match the specified size.")
        return

    # Read the elements of the list
    inputList = list(map(int, data[1:1 + inputList_size]))
    
    # Get the position of the smallest number
    result = funcSmallest(inputList)
    
    # Print the result
    print(result)
    
if __name__ == "__main__":
    main()
```
