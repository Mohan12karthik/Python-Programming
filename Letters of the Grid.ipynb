```python
def funcPuzzle(grid, words):
    def search_in_grid(word):
        rows, cols = len(grid), len(grid[0])
        len_word = len(word)

        def check_direction(start_row, start_col, direction_row, direction_col):
            r, c = start_row, start_col
            for i in range(len_word):
                if  r < 0 or r >= rows or c < 0 or c >= cols or grid[r][c] != word[i]:
                    return False
                r += direction_row
                c += direction_col
            return True
        
        for r in range(rows):
            for c in range(cols):
                # Check all 8 possible directions
                if (check_direction(r, c, 0, 1) or # Right
                    check_direction(r, c, 0, -1) or # Left
                    check_direction(r, c, 1, 0) or # Down
                    check_direction(r, c, -1, 0) or # Up
                    check_direction(r, c, 1, 1) or # Down-Right
                    check_direction(r, c, -1, -1) or # Up-Left
                    check_direction(r, c, 1, -1) or # Down-Left
                    check_direction(r, c, -1, 1)): # Up-Right
                    return True
        return False

    results = []
    for word in words:
        if search_in_grid(word):
            results.append("Yes")
        else:
            results.append("No")
    
    return ' '.join(results)

def main():
    import sys
#     input = sys.stdin.read
#     data = input().splitlines()
    
#     grid_rows, grid_cols = map(int, data[0].split())
#     grid = [line.split() for line in data[1:grid_rows+1]]
    
#     word_size = int(data[grid_rows+1])
#     words = data[grid_rows+2].split()
    
#     result = funcPuzzle(grid, words)
#     print(result)
        # Prompting user for input
    grid_rows, grid_cols = map(int, input("Enter the number of rows and columns of the grid (e.g., '3 3'): ").split())
    
    grid = []
    print(f"Enter the grid of characters row by row ({grid_rows} rows):")
    for _ in range(grid_rows):
        grid.append(input().split())

    word_size = int(input("Enter the number of words to search: "))
    words = input(f"Enter {word_size} words separated by space: ").split()

    result = funcPuzzle(grid, words)
    print("Result:", result)

if __name__ == "__main__":
    main()
```
