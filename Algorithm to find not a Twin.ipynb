```python
def funcTwins(inputArr):
    from collections import Counter
    
    # Count the frequency of each element
    count = Counter(inputArr)
    
    # Collect non-twin elements (frequency == 1)
    non_twins = [num for num, freq in count.items() if freq == 1]
    
    # Return the smallest non-twin element or -1 if there are none
    if non_twins:
        return min(non_twins)
    else:
        return -1

def main():
    import sys
#     input = sys.stdin.read
#     data = input().split()
    
#     # Read input size
#     inputArr_size = int(data[0])
    
#     # Read input array
#     inputArr = list(map(int, data[1:]))
    
#     result = funcTwins(inputArr)
#     print(result)

    # Prompting user for input
    inputArr_size = int(input("Enter the size of the array: "))

    # Reading input array
    inputArr = list(map(int, input(f"Enter {inputArr_size} integers separated by spaces: ").split()))

    # Calculate the result
    result = funcTwins(inputArr)
    
    # Print the result
    print(result)

if __name__ == "__main__":
    main()
```
